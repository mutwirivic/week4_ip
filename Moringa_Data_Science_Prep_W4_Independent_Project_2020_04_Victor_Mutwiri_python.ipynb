# Data cleaning

```
# Drop rows where city is not Paris, month is not April and year is not 2018 and assign the new dataframe to df1
df1 = df[~((df['City']!='Paris') | (df['year']!=2018) | (df['month']!=4))]
```

```
#Drop column 'Bluecar counter' because it is equivalent to column cars
df1 = df1.drop(['Bluecar counter'], axis =1)
```

```
df1
```

```
#Drop rows where charging status is future because there are no resources
df1 = df1[~(df1['Charging Status']=='future')]
df1
```

```
#Drop column 'displayed comment'
df1 = df1.drop(['Displayed comment'], axis =1)
```

```
#Drop rows where Kind = center because there are no resources
df1 = df1[~(df1['Kind']=='CENTER')]
```

```
#Drop ID column
df1 = df1.drop(['ID'], axis = 1)
```

```
#Drop column 'Geo point'
df1 = df1.drop(['Geo point'], axis =1)
```

```
df1['Rental status'].unique()
```

```
#Drop rows where rental status is not 'operational' or 'broken'
df1 = df1[((df1['Rental status']=='operational') | (df1['Rental status']=='broken'))]
```
